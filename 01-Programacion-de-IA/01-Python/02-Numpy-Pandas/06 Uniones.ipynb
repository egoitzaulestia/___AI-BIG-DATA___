# Ejercicios Uniones

Para este ejercicio se usarán los datos del archivo SalePrice.csv

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/SalePrice.csv')
```

## Dividir con registros repetidos y variables únicas
 Dividir el DataFrame en tres grupos con registros comunes (que unos se repitan en los grupos y otros no) y variables diferentes (que no se repitan)


```python
# Dividir el DataFrame en tres grupos con registros comunes (que unos se repitan en los grupos y otros no) y variables diferentes (que no se repitan)
df1 = df[['MSSubClass', 'MSZoning', 'LotFrontage']]
df2 = df[['LotArea', 'Street', 'Alley', 'LotShape']]
df3 = df[['OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd']]

df1 = df1[df1.index % 3 == 0]
df2 = df2[df2.index % 9 == 0]
df3 = df3[df3.index % 12 == 0]

df1 = df1.head(5)
df2 = df2.head(5)
df3 = df3.head(5)

# Mostrar las primeras filas de cada DataFrame para verificar la división
print("DataFrame 1:")
print(df1.head())

print("\nDataFrame 2:")
print(df2.head())

print("\nDataFrame 3:")
print(df3.head())
```

## Dividir con registros repetidos y variables repetidas
Dividir el DataFrame en tres grupos con registros comunes (que unos se repitan en los grupos y otros no) y variables comunes (que unas se repitan y otras no)


```python
df1_new = df[['Id', 'MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea']]
df2_new = df[['Id', 'LotArea', 'Street', 'Alley', 'OverallQual']]
df3_new = df[['Id', 'OverallQual', 'OverallCond', 'YearBuilt', 'LotFrontage']]


df1_new = df1_new[df1_new['Id'] % 3 == 0]
df2_new = df2_new[df2_new['Id'] % 9 == 0]
df3_new = df3_new[df3_new['Id'] % 12 == 0]

df1_new = df1_new.head(5)
df2_new = df2_new.head(5)
df3_new = df3_new.head(5)

# Mostrar las primeras filas de cada DataFrame para verificar la división
print("DataFrame 1:")
print(df1_new.head())

print("\nDataFrame 2:")
print(df2_new.head())

print("\nDataFrame 3:")
print(df3_new.head())
```

## Realizar uniones de los tres primeros grupos:


### a. Incluyendo todas las variables y todas las observaciones, por el índice.
Hacer una concatenación horizontal y otra vertical

```python
# Concatenación horizontal
pd.concat([df1, df2, df3], axis=1)
```

```python
# Concatenación vertical
pd.concat([df1, df2, df3])
```

### b. Incluyendo las observaciones comunes de las tres tablas y todas las variables, por el índice.
concatenación horizontal y vertical usando "inner"

```python
pd.concat([df1, df2, df3], axis=1, join='inner')
```

```python
pd.concat([df1, df2, df3], join='inner')
```

### c. Incluyendo las observaciones de la tabla dos y enriqueciendo esa información con la información de las otras dos tablas, por el índice.


```python
pd.concat([df1, df2, df3], axis=1)
```

```python
pd.concat([df1, df2, df3], axis=1).reindex(df2.index)
```

## Realizar dos grupos con observaciones comunes y variables únicas, incluyendo la Id en las dos tablas:


```python
df1 = df[['Id', 'MSSubClass', 'MSZoning', 'LotFrontage']]
df2 = df[['Id', 'LotArea', 'Street', 'Alley', 'OverallQual']]

df1 = df1[df1['Id'] % 3 == 0]
df2 = df2[df2['Id'] % 6 == 0]

df1 = df1.head(5)
df2 = df2.head(5)

# Mostrar las primeras filas de cada DataFrame para verificar la división
print("DataFrame 1:")
print(df1.head())

print("\nDataFrame 2:")
print(df2.head())
```

### a. Une las dos tablas usando la columna Id como clave de la unión

```python
# prompt: Une las dos tablas usando la columna Id como clave de la unión

pd.merge(df1, df2, on='Id', how = "outer")
```

### b. Incluyendo las observaciones comunes de las tres tablas y todas las
variables, por la Id.


```python
pd.merge(df1, df2, on='Id', how = "inner")
```

### c. Incluyendo las observaciones de la tabla dos y enriqueciendo esa
información con la información de la otra tabla, por la Id.


```python
pd.merge(df1, df2, on='Id', how = "right")
```
