```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# Cargar y desempaquetar los datos
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Nombres de las clases
class_names = ['Camiseta/Top', 'Pantalón', 'Jersey', 'Vestido', 'Abrigo',
               'Sandalia', 'Camisa', 'Zapatilla', 'Bolso', 'Bota']

# Verificar la forma de los datos
print("Forma de X_train:", X_train.shape)
print("Forma de y_train:", y_train.shape)
print("Forma de X_test:", X_test.shape)
print("Forma de y_test:", y_test.shape)

# Visualizar una imagen de ejemplo
plt.figure()
plt.imshow(X_train[0], cmap='gray')
plt.colorbar()
plt.title(f'Clase: {class_names[y_train[0]]}')
plt.show()
```

```python
# 1. Normalizar los datos
X_train = X_train / 255.0
X_test = X_test / 255.0

# 2. Crear el modelo CNN
model = keras.Sequential([
    # Reshape para añadir canal (necesario para CNN)
    keras.layers.Reshape((28, 28, 1), input_shape=(28, 28)),
    
    # Primera capa convolucional
    keras.layers.Conv2D(32, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    
    # Segunda capa convolucional
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    
    # Aplanar para las capas densas
    keras.layers.Flatten(),
    
    # Capas densas
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10, activation='softmax')
])

# Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Mostrar el resumen del modelo
model.summary()

# 3. Entrenar el modelo
history = model.fit(X_train, y_train, 
                    epochs=10, 
                    batch_size=32,
                    validation_split=0.2)

# 4. Evaluar el modelo con datos de prueba
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"\nPrecisión en el conjunto de prueba: {test_accuracy:.4f}")

# 5. Visualizar el entrenamiento
plt.figure(figsize=(12, 4))

# Gráfica de precisión
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del modelo')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()

# Gráfica de pérdida
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida del modelo')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()

plt.show()

# 6. Hacer algunas predicciones y visualizarlas
predictions = model.predict(X_test[:5])

plt.figure(figsize=(15, 3))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(X_test[i], cmap='gray')
    pred_label = class_names[np.argmax(predictions[i])]
    true_label = class_names[y_test[i]]
    plt.title(f'Pred: {pred_label}\nReal: {true_label}')
    plt.axis('off')
plt.show()
```

```python
# Importar las librerías necesarias
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Interacciones sociales
# 15 nodos y una probabilidad de conexión del 30%
G = nx.erdos_renyi_graph(n=15, p=0.3, seed=42)

# Etiquetas a los nodos para simular nombres de usuarios
for i, node in enumerate(G.nodes):
    G.nodes[node]['label'] = f'User_{i+1}'

# Calcular métricas de centralidad
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

# DataFrame
centrality_df = pd.DataFrame({
    'Node': [G.nodes[node]['label'] for node in G.nodes],
    'Degree Centrality': degree_centrality.values(),
    'Betweenness Centrality': betweenness_centrality.values(),
    'Closeness Centrality': closeness_centrality.values()
}).sort_values(by='Degree Centrality', ascending=False)

# Métricas
print("Métricas de centralidad:\n")
print(centrality_df)
```

```python
# Diseño del gráfico
pos = nx.spring_layout(G, seed=42)  # Diseño por fuerza de resorte

# Nodos y conexiones
plt.figure(figsize=(10, 8))
nodes = nx.draw_networkx_nodes(G, pos, 
                             node_color=list(degree_centrality.values()),
                             cmap=plt.cm.viridis, 
                             node_size=800)

nx.draw_networkx_edges(G, pos, 
                      alpha=0.5, 
                      edge_color="gray")

nx.draw_networkx_labels(G, pos, 
                       labels={node: G.nodes[node]['label'] for node in G.nodes},
                       font_size=10, 
                       font_color="white")

# Barra de centralidad
cbar = plt.colorbar(nodes, label="Degree Centrality")

# Título
plt.title("Interacciones Sociales", fontsize=14)
plt.axis("off")
plt.show()
```
